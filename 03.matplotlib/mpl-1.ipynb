```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
unrate = pd.read_csv('UNRATE.csv', parse_dates=['DATE'])
unrate.head()
```

```python
_ = plt.plot()
```

```python
first_twelve = unrate[0 : 12]
_ = plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
```

```python
_ = plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
_ = plt.xticks(rotation=35)
```

```python
_ = plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
_ = plt.xticks(rotation=45)

plt.xlabel('Month')
plt.ylabel('Unemployment count')
_ = plt.title('Monthly Unemployment Trends, 1948')
```
