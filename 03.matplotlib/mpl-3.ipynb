```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] =  (12, 4)

sns.set_style('darkgrid')
```

```python
plt.rcParams.get('figure.figsize')
```

```python
reviews = pd.read_csv('fandango_scores.csv')
reviews.head()
```

```python
ax = reviews['RT_norm_round'].value_counts().sort_index().plot.bar(title='RT_norm Bar')
ax.set_title('RT_norm', fontsize=15)
_ = ax.set_xlabel('RT_norm_round', fontsize=12)
_ = ax.set_ylabel('Counts', fontsize=12)
_ = plt.xticks(rotation=55)
_ = plt.yticks(rotation=55)
```

```python
_ = plt.figure(figsize=(7,7))

ax = sns.scatterplot(reviews['Fandango_Ratingvalue'], reviews['RT_user_norm'], alpha=0.6)
_ = ax.set_xlabel('Fandango')
_ = ax.set_ylabel('Rotten Tomatoes')
_ = ax.set_xlim(0, 5)
_ = ax.set_ylim(0, 5)
```
